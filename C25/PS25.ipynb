```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hepatitis_csv.csv')
```

```python
df.head()
```

```python
male_df = df[df['sex'] == 'male']
female_df = df[df['sex'] == 'female']
```

```python
male_df.head()
```

```python
female_df.head()
```

```python
merged_df = pd.concat([male_df, female_df], ignore_index=True)
```

```python
sorted_df = df.sort_values(by=['age', 'sgot', 'protime'], ascending=[True, True, True])
```

```python
sorted_df.head()
```

```python
transposed_df = df.T
```

```python
transposed_df.head()
```

```python
melted_df = pd.melt(df, id_vars=['age', 'sex'], var_name='Attribute', value_name='Value')
```

```python
melted_df.head()
```

```python
# Pivoting back to wide format based on Attribute
wide_df = melted_df.pivot_table(index=['age', 'sex'], 
                                 columns='Attribute', 
                                 values='Value', 
                                 aggfunc='first').reset_index()
```
