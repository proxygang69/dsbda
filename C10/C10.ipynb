```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
Setosa = df[df['variety'] == 'Setosa']
Versicolor = df[df['variety'] == 'Versicolor']
Virginica = df[df['variety'] == 'Virginica']
```

```python
Setosa.head()
```

```python
merged_df = pd.concat([Setosa, Versicolor], ignore_index=True)
merged_df
```

```python
sorted_df = df.sort_values(by='petal.length')
sorted_df.head()
```

```python
transposed_df = df.T
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['variety'], var_name='features', value_name='value')
melted_df
```

```python
wide_df = melted_df.pivot_table(index = 'variety', columns = 'features', values = 'value', aggfunc = 'mean').reset_index()
wide_df
```
