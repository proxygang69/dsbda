```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adult_dataset.csv')
```

```python
df.head()
```

a. Histogram
🔍 Objective:
Visualize the distribution of age across the dataset to understand the spread and common age ranges among individuals.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

b. Pie Chart
🔍 Objective:
Show the proportion of individuals in each income category (<=50K or >50K).

```python
income_counts = df['income'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Income Category Distribution')
plt.axis('equal')
plt.show()
```

c. Box Plot
🔍 Objective:
Compare the age distributions of individuals based on their income category to detect outliers and medians.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='income', y='age', data=df)
plt.title('Box Plot of Age by Income Category')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

d. Scatter Plot
🔍 Objective:
Visualize the relationship between age and hours-per-week to explore working patterns by age.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='hours-per-week', hue='income')
plt.title('Age vs Hours-per-Week by Income')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.show()
```

e. Add Boxplots to a Scatterplot
🔍 Objective:
Use a jointplot to show the scatter distribution of age vs hours-per-week, while also visualizing the marginal boxplots to understand distributions on each axis.

```python
sns.boxplot(data=df, x='income', y='hours-per-week', whis=1.5)
sns.stripplot(data=df, x='income', y='hours-per-week', color='black', alpha=0.3)
plt.title('Boxplot with Scatter Overlay: Hours-per-Week by Income')
plt.show()
```
