```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.meesho.com/banarasi-sarees-new-collection-banarasi-silk-sarees-2024-new-design-party-wear-silk-saree/p/7ky2ve')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('span', {'class': 'sc-eDvSVe gUjMRV Comment__CommentText-sc-1ju5q0e-3 cfdxfJ Comment__CommentText-sc-1ju5q0e-3 cfdxfJ'});
for review in reviews:
    print(review.get_text(), "\n")
```

```python
#average overall rating
ratings = soup.find('h1',{'class':'sc-eDvSVe cdZTwf'}).get_text();
print(ratings)
```

```python
#individual ratings

individual_ratings = soup.findAll('span',{'class':'sc-cjibBx eBDae'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text() + "\n");
```

```python
customer_name = soup.findAll('span',{'class':'sc-eDvSVe dugLmN'});
for cust_name in customer_name:
    print(cust_name.get_text())
```

```python
data = []

# Loop through all items (assumes lengths are same or aligned)
for i in range(len(reviews)):
    try:
        review_text = reviews[i].text.strip()
        rating = individual_ratings[i].text.strip()
        name = customer_name[i].text.strip()

        data.append({
            "Customer Name": name,
            "Rating": rating,
            "Review": review_text
        })
    except IndexError:
        # In case of mismatch in lengths
        continue
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.head()
```
